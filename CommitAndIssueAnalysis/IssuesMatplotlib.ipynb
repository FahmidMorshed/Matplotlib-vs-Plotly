```python
from github import Github
g = Github("access_token")
g = Github("b2fd9830b931a8281747a3244c6a101da6ebbf75")

repo=g.get_repo("matplotlib/matplotlib")

paginated_obj=repo.get_issues(state="closed")
```

```python
date=[]
def appendDate(paginated_obj):
    for singleIssue in paginated_obj:
        dateTime=singleIssue.created_at
        date.append(dateTime)
appendDate(paginated_obj)
appendDate(repo.get_issues(state="open"))
```

```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
```

```python
print(date)
```

```python
import numpy as np
df = pd.DataFrame(np.array(date))
print(df)
```

```python
# df=df.groupby(pd.Grouper(freq='M'))
# df=df.groupby(by=[b.index.month, b.index.year])
df['new'] = pd.Series([0 for x in range(len(df.index))], index=df.index)
df['new']=1
df.columns=['timestamp','new']
```

```python
df
```

```python
aggregate_data=df.set_index('timestamp').groupby(pd.Grouper(freq='M'))['new'].sum().reset_index()
aggregate_data
```

```python
x = aggregate_data['timestamp']
y = aggregate_data['new']

```

```python
# plot
plt.plot(x,y,color="red")
plt.xlabel('Year')
plt.ylabel('Number of Issues')
plt.title('Issues vs Timeline Chart')
# beautify the x-labels
plt.gcf().autofmt_xdate()
plt.show()
```
