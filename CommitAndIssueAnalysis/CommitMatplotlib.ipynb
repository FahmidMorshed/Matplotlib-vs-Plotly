```python
repo=g.get_repo("plotly/plotly.py")

paginated_obj=repo.get_commits()
date=[]
for singleCommit in paginated_obj:
    gitCommitObj=singleCommit.commit
    gitAuthor=gitCommitObj.author
    dateTime=gitAuthor.date
    date.append(dateTime)
    
df = pd.DataFrame(np.array(date))
df['new'] = pd.Series([0 for x in range(len(df.index))], index=df.index)
df['new']=1
df.columns=['timestamp','new']
aggregate_data=df.set_index('timestamp').groupby(pd.Grouper(freq='M'))['new'].sum().reset_index()
m = aggregate_data['timestamp']
k = aggregate_data['new']
```

```python
# plot
plt.plot(x,y)
plt.plot(m,k)

plt.legend(['Matplotlib', 'Plotly'], loc='upper left')

plt.xlabel('Year')
plt.ylabel('Number of Commits')
plt.title('Contributions vs Timeline Chart')
# beautify the x-labels
plt.gcf().autofmt_xdate()
plt.show()




```
